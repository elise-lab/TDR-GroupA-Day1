```python
def find_theta4(w2):
    distance = norm2(w2, [1,0])
    
    gamma = np.arcsin(w2[1]/distance)
    phi = np.arccos((w2[0]-1)/distance)
    
    if gamma == phi:
        theta4 = gamma
    elif (w2[0] < 1) and (w2[0]>1):
        theta4 = phi
    elif (w2[0]<1):
        theta4 = np.pi - gamma
    else:
        theta4 = 2*np.pi + gamma
        
    return theta4
    
```

```python
def generate_pentagons(e1, e2, e3, e4, theta, n): #this is the latest version
    w1 = [e1*np.cos(theta), e1*np.sin(theta)]
    c = e3+e4
    b = norm2(w1, [1,0])
    
    if (e1)>(e2+e3+e4+1):
        return("stupid")
    
    if (e3+e4)>(b+e2):
        min_theta2=0
        max_theta2=2*np.pi #if literally any angle will work, the law of cosines breaks down
        #it doesn't account for the fact that the "side" can become shorter if it isn't a straight line
    else:


        arccos_input = (e2**2 + b**2 - c**2)/(2*e2*b)

        #print(b)
        #print(c)
        #print(arccos_input)

        if abs(arccos_input) > 0.999:
            return "stupid"

        max_theta2 = np.arccos((e2**2 + b**2 - c**2)/(2*e2*b))

        theta_int = np.arcsin((np.sin(theta)*e1)/b)

        min_theta2 = -max_theta2 - theta_int
        max_theta2 += -1*theta_int
        
    print("min, max")
    print(min_theta2)
    print(max_theta2)
    
    coords= []
    
    for i in range(n):

        theta_2 = np.random.uniform(min_theta2,max_theta2) #generate a second random angle, within the prescribed bounds
        w2 = [(e2*np.cos(theta_2) + w1[0]), (e2*np.sin(theta_2) + w1[1])]


        distance = norm2(w2, [1,0])
        print("arccos input")
        print((e4**2 - e3**2 + distance**2)/(2*e4*distance))
        theta_3 = np.arccos((e4**2 - e3**2 + distance**2)/(2*e4*distance)) #angle from (1,0) to w3

        theta_4  = find_theta4(w2) #angle from (1,0) to w2
        theta_5a = theta_4+theta_3
        theta_5b = theta_4-theta_3


        w3a = [1+e4*np.cos(theta_5a), e4*np.sin(theta_5a)]
        w3b = [1+e4*np.cos(theta_5b), e4*np.sin(theta_5b)]

        coords_a = []
        coords_a.extend(w1)
        coords_a.extend(w2)
        coords_a.extend(w3a)

        coords_b = []
        coords_b.extend(w1)
        coords_b.extend(w2)
        coords_b.extend(w3b)
        
        coords.append(coords_a)
        coords.append(coords_b)
    
    
    
    return (coords)
```
