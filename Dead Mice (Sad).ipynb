```python
import pandas as pd
import numpy as np
from scipy.spatial import distance

#topological data analysis
from ripser import ripser
from persim import plot_diagrams

#plotting and visualization
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from matplotlib.patches import Circle

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rbc_p2 = pd.read_excel("mice_data.xlsx", skiprows = 1, usecols = [22,43,120], header=None, index_col=None)
data = rbc_p2.to_numpy()

#nkg7 --> 24
#fech --> 68, 69
#rbc --> 4
#"oops all ones" --> 120
```

```python
fig = go.Figure(data=[go.Scatter3d(
    x=data[:,0], y=data[:,1], z=data[:,2], 
    # x=data[;,1], y=data[:,2], z=data[:,0], #if you want to graph time 
    mode ='markers', 
    marker=dict(size = 3 , color = 'grey'))])

fig.update_layout(scene= dict(zaxis = dict(range=[-5, 5])))
fig.show()
```

```python
rips_persistence = ripser(data, n_perm = 75, coeff = 2, maxdim=2)
dgms = rips_persistence['dgms']
idx_perm = rips_persistence['idx_perm']
plt.figure(figsize = (4,4)) 
plot_diagrams(dgms)
```

```python
def max_per2(rips, n=1):
    max_diff = 0
    for i in range(len(rips['dgms'][n])):
        diff = rips['dgms'][n][i][1] - rips['dgms'][n][i][0]
        if diff > max_diff:
            max_diff = diff
            #print(max_diff)
    
    return max_diff
```

```python
current_max= [0,0]
second_place = [0,0]
third_place = [0,0]

for i in range(105):
    rbc_p2 = pd.read_excel("mice_data.xlsx", skiprows = 1, usecols = [4,(i+12), (i+13), (i+14), (i+15), (i+16)], header=None, index_col=None)
    data = rbc_p2.to_numpy()
    
    rips_persistence = ripser(data, n_perm = 75, coeff = 2, maxdim=2)
    
    max_persistence=max_per2(rips_persistence)
    if max_persistence > current_max[1]:
        third_place = second_place
        second_place = current_max
        current_max = [(i+12), max_persistence]
    
    print("Max persistence for column starting at " +str(i+12)+": "+str(max_persistence))

    
print(current_max)
print(second_place)
print(third_place)
```

```python
from sklearn.decomposition import PCA
import plotly.graph_objects as go
from plotly.subplots import make_subplots

```

```python
pca = PCA(n_components=3)
SW_pca = pca.fit(data2).transform(data2)

print('Explained Variance', pca.explained_variance_ratio_)

fig = go.Figure(data=[go.Scatter3d(
    x=SW_pca[:,0], y=SW_pca[:,1], z=SW_pca[:,2], 
    mode ='markers', 
    marker=dict(size = 2, color = 'grey')
)] , layout = go.Layout(scene = dict(aspectmode= 'data')))

fig.show()

print(SW_pca.shape)
```

```python
alpha = 3.25
distMat = distance.squareform(distance.pdist(SW_pca))

ii = []; jj = []; kk = []
e_x =[]; e_y =[]; e_z =[]

for i in range(75):
    for j in range(i+1,75):
        if distMat[i,j] < alpha:
            # add edge (i,j)
            e_x.extend([SW_pca[i,0], SW_pca[j,0], None])
            e_y.extend([SW_pca[i,1], SW_pca[j,1], None])
            e_z.extend([SW_pca[i,2], SW_pca[j,2], None])
            
            for k in range(j+1,75):
                if np.max([distMat[j,k], distMat[i,k]]) < alpha:
                    # add triangle (i,j,k)
                    ii.append(i); jj.append(j); kk.append(k)
                
vertices = go.Scatter3d(mode = 'markers', name = 'vertices', 
                        x = SW_pca[:,0], y = SW_pca[:,1],  z = SW_pca[:,2], 
                        marker=dict(size = 3 , color = 'grey'))

edges = go.Scatter3d(mode='lines', name = 'edges', 
                     x=e_x, y=e_y, z=e_z, 
                     line=dict(color= 'rgb(70,70,70)', width=1))  

triangles = go.Mesh3d(x=SW_pca[:,0], y=SW_pca[:,1], z=SW_pca[:,2],  i = ii, j = jj, k = kk,  color='lightpink', opacity=0.2)

fig = go.Figure(data=[vertices, edges, triangles])
fig.update_traces(hoverinfo="none")
fig.update_layout(scene= dict(
                      xaxis = dict(showspikes=False), 
                      yaxis = dict(showspikes=False), 
                      zaxis = dict(showspikes=False,range=[-5, 5])))
fig.show()
```
