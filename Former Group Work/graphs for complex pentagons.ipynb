```python
import matplotlib.pyplot as plt
import numpy as np

# Example set of complex numbers
complex_numbers = [
    np.cos(2*np.pi/5) + np.sin(2*np.pi/5)*1j,
    np.cos(4*np.pi/5) + np.sin(4*np.pi/5)*1j,
    np.cos(6*np.pi/5) + np.sin(6*np.pi/5)*1j,
    np.cos(8*np.pi/5) + np.sin(8*np.pi/5)*1j,
    1
]

# Complex numbers a and b
a = (1 + 3j/np.abs(1 + 3j))
b = -1 + 2j
# Multiply each element by a and add b (Modified complex numbers)
complex_numbers_modified = [z*a for z in complex_numbers]
complex_numbers_sim = [z*a+b for z in complex_numbers]


# Sort the complex numbers based on their angles (Unmodified complex numbers)
complex_numbers_sorted = sorted(complex_numbers, key=np.angle)

# Separate the real and imaginary parts of the modified complex numbers
real_parts_modified = [z.real for z in complex_numbers_modified]
imaginary_parts_modified = [z.imag for z in complex_numbers_modified]

# Separate the real and imaginary parts of the unmodified complex numbers
real_parts_sorted = [z.real for z in complex_numbers_sorted]
imaginary_parts_sorted = [z.imag for z in complex_numbers_sorted]

#with a and b 
real_parts_sim_sort = [z.real for z in complex_numbers_sim]
imaginary_parts_sim_sort = [z.imag for z in complex_numbers_sim]

# Plot the modified complex numbers on the complex plane with lines connecting the dots
plt.plot(real_parts_modified, imaginary_parts_modified, '-o', color = 'orange', label='Modified')

# Plot the unmodified complex numbers on the complex plane with lines connecting the dots
plt.plot(real_parts_sorted, imaginary_parts_sorted, '-o', color='blue', label='Unmodified' )

#plot fully translated pentagon
plt.plot(real_parts_sim_sort, imaginary_parts_sim_sort, '-o', color='green', label='a and b')

# Set fixed axis limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.title("Complex Pentagons With Rotation and Translation")
plt.xticks(np.arange(-5, 6, 1))
plt.yticks(np.arange(-5, 6, 1))
# Set custom tick locations and labels for x-axis and y-axis
```
